```python
# import all dependencies
import pandas as pd 
from citipy import citipy
import numpy as np

import requests 
from config import weather_api_key
```

```python
# create set of 2000 random lat and lng
# generate random lat 
lats = np.random.uniform(low=-90.000, high = 90.000, size = 2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size = 2000)

# create zip object in memory (packing the two together)
lat_lngs = zip(lats, lngs)
# packed into tuple (1500 of them) and access as 0,1 index form
lat_lngs

# add lat and lng to list 
coordinates = list(lat_lngs)
```

```python
# get nearest city using citypy module

# Create a list for holding the cities.
cities = []
# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
# Print the city count to confirm sufficient count.
#print(cities)
len(cities)
```

```python
# perform api call with openweathermap 
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
print(url)
```

```python
# get lat and lng, max temp, % humiditiy, % cloudiness, wind speed, weather description
city_data = []
print("Beginning Data Retrieval           ")
print("-----------------------------------")

# Create a counters 
record_count = 1
set_count = 1

# fixed version that adds not city index, but name 
# enumerate helps refrence the index and city in list
for index, city in enumerate(cities):
    
    # Group cities in sets of 50 for logging purposes.
    if(index % 50 == 0 and index >= 50):
        set_count += 1
        record_count = 1
        
    # Create endpoint URL with each city.
    city_url = f"{url}&q={city.replace(' ', '+')}"
    
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    record_count += 1
    
    try:
        city_weather = requests.get(city_url).json()
        # Parse out the max temp, humidity, and cloudiness, description
        city_lat = city_weather['coord']['lat']
        city_lng = city_weather['coord']['lon']
        city_max_temp = city_weather['main']['temp_max']
        city_humidity = city_weather['main']['humidity']
        city_clouds = city_weather['clouds']['all']
        city_wind = city_weather['wind']['speed']
        city_country = city_weather['sys']['country']
        city_weather_description = city_weather['weather'][0]['description']
        # Append the city information into city_data.
        city_data.append(
        {
            'City': city,
            'Lat': city_lat,
            'Lng': city_lng,
            'Max Temp': city_max_temp,
            'Current Description': city_weather_description,
            'Humidity': city_humidity,
            'Cloudiness': city_clouds,
            'Wind Speed': city_wind,
            'Country': city_country
        })
    except Exception as e:
        # print('City not found. Skipping...')
        print(e)
        pass
# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# create new df with city, country, lat, lng, max temp, humidity
# cloudiness, wind speed, current description
newcity_df = pd.DataFrame(city_data)
newcity_df.head(10)
```

```python
# change the order to match df wants from module
reorderCol =  ['City', 'Country', 'Lat', 'Lng', 'Max Temp', 'Humidity', 
                    'Cloudiness', 'Wind Speed', 'Current Description']
newcity_df = newcity_df[reorderCol]
newcity_df.head(10)
```

```python
# Create the output file (CSV).
output_data_file = "WeatherPy_Database.csv"

# Export the newcity_df into a CSV.
newcity_df.to_csv(output_data_file, index_label="City_ID")
```
